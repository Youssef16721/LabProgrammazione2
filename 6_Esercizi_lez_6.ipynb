### **2. Fitta i dati**
Prova a fittare i dati sottostante con le migliori curve, calcola il MAE e l'RMSE

```python
import numpy as np
temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
import matplotlib.pyplot as plt
months = np.arange(12)
plt.plot(months, temp_max, "ro")
plt.plot(months, temp_min, "bo")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")
```

```python
from sklearn.metrics import mean_absolute_error,root_mean_squared_error
coffmax = np.polyfit(months,temp_max, deg=4) 
coffmin = np.polyfit(months,temp_min, deg=4) 
pred_max = np.poly1d(coffmax)
pred_min = np.poly1d(coffmin)

fit_max = pred_max(months)
fit_min = pred_min(months)

mae_max = mean_absolute_error(temp_max,fit_max)
rmse_max= root_mean_squared_error(temp_max, fit_max)
mae_min= mean_absolute_error(temp_min,fit_min)
rmse_min=root_mean_squared_error(temp_min,fit_min)
print("MAE max:", mae_max, "RMSE max:", rmse_max)
print("MAE min:", mae_min, "RMSE min:", rmse_min)

plt.plot(months,temp_max,"o",label="Original Data")
plt.plot(months,temp_min, "o",label="Original Data")
plt.plot(months,fit_max,label="Polinomial Data MAX")
plt.plot(months,fit_min,label="Polinomial Data MIN")
plt.legend()
plt.show()
```

### **3. Modello di regressione lineare dei seguenti dati**

 Calcola un modello di regressione lineare delle colonne mpg e disp del famoso dataset mtcars.

Dove: 

mpg = Miles Per Gallon → miglia per gallone, cioè una misura del consumo di carburante. Più alto è il valore, più efficiente è l’auto.

disp = Displacement → cilindrata del motore, in pollici cubici (cubic inches). Rappresenta il volume totale dei cilindri del motore. Più è alto, maggiore è la potenza potenziale del motore (ma anche il consumo).

```python
import pandas as pd
import scipy as sp

# Load dataset from URL
df = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')

# Display the dataframe
df
```

```python
import seaborn as sns
x = df['disp']
y = df['mpg']

coff = np.polyfit(x,y,deg=1)
lm = np.poly1d(coff)
y_pred = lm(x)

sns.scatterplot(x='disp', y='mpg', data=df, color='blue', label='Dati reali')
plt.plot(x,y_pred,linestyle="-",label="LM")
plt.title('Regressione Lineare: mpg ~ disp')
plt.xlabel('Cilindrata (disp)')
plt.ylabel('Miglia per Gallone (mpg)')
plt.legend()
plt.grid(True)
plt.show()

```
